## 1. Fondasi Filosofis EDA: Melampaui Angka

John Tukey (1977), bapak dari EDA, mendefinisikan disiplin ini sebagai sikap terhadap analisis data yang mengutamakan intuisi, fleksibilitas, dan penemuan. 
Berbeda dengan Confirmatory Data Analysis (CDA) yang berfokus pada pengujian hipotesis yang sudah ditentukan sebelumnya (p-values, t-tests), 
EDA memungkinkan data untuk "berbicara" sendiri tanpa dibatasi oleh asumsi kaku.

### Mengapa EDA Adalah Kunci Keberhasilan ML?
- **Verifikasi Kualitas Data**: EDA membantu kita menemukan kesalahan sensor, bias dalam pengambilan sampel, atau missing values yang tidak terduga.
- **Validasi Asumsi Model**: Sebagian besar algoritma statistik (seperti Regresi Linear) mengasumsikan distribusi normal dan homoskedastisitas. EDA memverifikasi hal ini.
- **Feature Engineering & Selection**: Melalui visualisasi, kita sering menemukan ide untuk menggabungkan dua variabel menjadi satu fitur baru yang lebih prediktif.
- **Komunikasi Stakeholder**: Grafik yang dihasilkan selama EDA seringkali menjadi cara terbaik untuk menjelaskan temuan awal kepada tim bisnis.

Dalam dunia nyata, 80% waktu seorang Data Scientist dihabiskan untuk EDA dan preprocessing. Sisanya barulah pemodelan. Jika fondasi EDA Anda lemah, model ML Anda akan menderita fenomena *Garbage In, Garbage Out*.

### Paradigma Modern: EDA 2.0
Di era big data, EDA telah berkembang menjadi proses semi-otomatis menggunakan alat seperti Sweetviz, Pandas Profiling, atau D-Tale. Namun, pemahaman manual tetap krusial untuk melakukan kustomisasi analisis yang tidak bisa ditangkap oleh alat otomatis tersebut.

## 2. Estimasi Lokasi (Estimates of Location) - Analisis Mendalam

Lokasi adalah nilai pusat di mana sebagian besar data berada. Namun, definisi "pusat" bisa bervariasi tergantung pada distribusi data Anda.

### A. Mean (Arithmetic Mean)
Rata-rata sederhana yang kita kenal sejak sekolah. Sangat efisien secara statistik tetapi sangat rentan terhadap pencilan (outliers).
$$\bar{x} = \frac{\sum x_i}{n}$$

### B. Trimmed Mean (Truncated Mean)
Solusi cerdas untuk menangani outliers tanpa menghapusnya secara manual. Kita memotong sebagian persentase tertentu dari ujung atas dan bawah distribusi.
$$\bar{x}_{trim} = \frac{\sum_{i=p+1}^{n-p} x_{(i)}}{n-2p}$$

### C. Weighted Mean
Digunakan ketika beberapa observasi dianggap lebih penting atau representatif daripada yang lain.
$$\bar{x}_w = \frac{\sum w_i x_i}{\sum w_i}$$

### D. Median
Nilai tengah yang membagi data menjadi dua bagian yang sama besar. Median adalah statistik **robust** karena tidak terpengaruh oleh seberapa besar nilai ekstrem di ujung distribusi.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

def simulate_location_data(n_samples=1000, n_outliers=10):
    """
    Melakukan simulasi pembuatan dataset yang mengandung data normal dan outliers ekstrem.
    
    Fungsi ini dirancang untuk mendemonstrasikan bagaimana statistik lokasi yang berbeda 
    bereaksi terhadap kehadiran pencilan (outliers). Kita akan membuat data dasar dari 
    distribusi normal, lalu menyuntikkan beberapa nilai yang sangat jauh dari pusat.
    
    Langkah-langkah detail:
    1. Menentukan seed random untuk reproduksibilitas (seed=42).
    2. Membuat array data 'clean' dengan distribusi normal (mean=100, std=15).
    3. Membuat array data 'outliers' dengan distribusi uniform (rentang 1000 - 5000).
    4. Menggabungkan kedua array tersebut menjadi satu dataset tunggal.
    5. Mengacak urutan data agar outliers tidak hanya berada di akhir array.
    
    Parameter:
    ----------
    n_samples : int, default=1000
        Jumlah sampel data normal yang akan dihasilkan menggunakan np.random.normal.
    n_outliers : int, default=10
        Jumlah data pencilan ekstrem yang akan ditambahkan ke dataset.
        
    Returns:
    -------
    data : numpy.ndarray
        Array gabungan yang berisi data normal dan data pencilan.
        
    Penjelasan Teknis:
    -----------------
    Data normal dihasilkan dengan rata-rata 100 dan deviasi standar 15.
    Outliers dihasilkan secara acak dalam rentang 1000 hingga 5000.
    Penggabungan dilakukan menggunakan np.concatenate untuk memastikan struktur array yang efisien.
    """
    np.random.seed(42)
    clean_data = np.random.normal(loc=100, scale=15, size=n_samples)
    outliers = np.random.uniform(low=1000, high=5000, size=n_outliers)
    final_data = np.concatenate([clean_data, outliers])
    np.random.shuffle(final_data)
    return final_data

# Eksekusi simulasi dengan parameter custom
mixed_data = simulate_location_data(n_samples=2000, n_outliers=20)

# Penghitungan statistik menggunakan berbagai library
mean_val = np.mean(mixed_data)
median_val = np.median(mixed_data)
trimmed_mean_10 = stats.trim_mean(mixed_data, 0.1)

print(f"--- Analisis Lokasi Mendalam ---")
print(f"Mean (Rata-rata): {mean_val:.4f}")
print(f"Median (Nilai Tengah): {median_val:.4f}")
print(f"Trimmed Mean (Potongan 10%): {trimmed_mean_10:.4f}")
print(f"Selisih Mean vs Median: {abs(mean_val - median_val):.4f}")

# Visualisasi High-Definition
plt.figure(figsize=(15, 8))
sns.set_style("whitegrid")
sns.kdeplot(mixed_data, fill=True, color="royalblue", label="Kepadatan Data", alpha=0.4)
plt.axvline(mean_val, color='crimson', linestyle='--', linewidth=2.5, label=f'Mean: {mean_val:.1f}')
plt.axvline(median_val, color='forestgreen', linestyle='-', linewidth=2.5, label=f'Median: {median_val:.1f}')
plt.axvline(trimmed_mean_10, color='darkorchid', linestyle=':', linewidth=2.5, label=f'Trimmed Mean: {trimmed_mean_10:.1f}')

plt.title("Dampak Outliers pada Lokasi Distribusi", fontsize=18, fontweight='bold')
plt.xlabel("Nilai Observasi", fontsize=14)
plt.ylabel("Kepadatan Relatif", fontsize=14)
plt.xlim(0, 400) # Fokus pada area di mana data mayoritas berada
plt.legend(fontsize=12)
plt.annotate('Outliers menarik mean ke kanan',
             xy=(mean_val, 0.005), xytext=(200, 0.015),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize=12)

plt.show()
```

### Interpretasi Visualisasi Lokasi

Grafik di atas menunjukkan fenomena klasik dalam statistik: pergeseran rata-rata (mean) akibat nilai ekstrem. Meskipun kita hanya menambahkan 20 pencilan (sekitar 1% dari total data), nilai **Mean** tertarik cukup jauh ke arah kanan. Dalam simulasi ini, Mean berada di angka yang signifikan lebih tinggi daripada puncak kurva kepadatan mayoritas data. Ini menunjukkan bahwa Mean memberikan representasi yang bias jika data mengandung pencilan atau memiliki ekor yang berat (heavy tails).

Sebaliknya, **Median** tetap kokoh dan sangat dekat dengan puncak kurva (mode). Median hanya memedulikan urutan nilai, bukan besaran nilai tersebut. Oleh karena itu, median adalah pilihan terbaik ketika Anda berurusan dengan data yang memiliki varians tinggi atau kesalahan sensor yang sering terjadi. Dalam konteks ekonomi, median gaji seringkali ribuan dolar lebih rendah daripada mean gaji karena keberadaan segelintir individu dengan pendapatan sangat tinggi.

**Trimmed Mean** (dalam hal ini membuang 10% data terkecil dan terbesar) memberikan hasil yang sangat menarik. Ia hampir berhimpit dengan median tetapi secara teoritis lebih efisien jika data bersih mengikuti distribusi normal. Trimmed mean memberikan "keamanan" ekstra: ia tidak sebias mean, tetapi tidak membuang informasi sebanyak median. Ini adalah teknik favorit dalam kompetisi matematika atau penilaian olahraga di mana skor juri tertinggi dan terendah sering dibuang untuk menghindari bias subjektif.

## 3. Estimasi Variabilitas (Estimates of Variability)

Variabilitas memberi tahu kita seberapa jauh titik data menyebar dari pusat. Tanpa memahami variabilitas, kita tidak bisa mengukur ketidakpastian atau risiko.

### A. Variance dan Standard Deviation
Standard deviation adalah akar kuadrat dari varians. Ia memberikan skala yang sama dengan data asli. Namun, karena menggunakan kuadrat selisih, nilai besar (outliers) akan mendominasi hasil akhir.
$$\sigma = \sqrt{\frac{\sum (x_i - \bar{x})^2}{n-1}}$$

### B. Interquartile Range (IQR)
Selisih antara persentil ke-75 (Q3) dan persentil ke-25 (Q1). Ini adalah ukuran penyebaran yang sangat stabil dan sering digunakan untuk mendeteksi outlier secara visual melalui boxplots.

### C. Median Absolute Deviation (MAD)
Ini adalah versi robust dari standar deviasi. Dihitung dengan mencari median dari selisih absolut setiap titik terhadap median utama.
$$\text{MAD} = \text{median}(|x_i - 	ext{median}(X)|)$$

### D. Range (Rentang)
Hanya selisih antara nilai maksimum dan minimum. Sangat sensitif terhadap outliers sehingga jarang digunakan dalam analisis serius kecuali sebagai ringkasan cepat.

```python
def calculate_variability_metrics(data):
    """
    Menghitung berbagai metrik variabilitas untuk membandingkan metode sensitif vs robust.
    
    Fungsi ini melakukan kalkulasi mendalam terhadap penyebaran data. Variabilitas adalah 
    aspek krusial dalam EDA karena menentukan lebar interval kepercayaan dan reliabilitas 
    estimasi titik kita.
    
    Detail Operasi:
    1. Menghitung Standard Deviation menggunakan np.std dengan ddof=1 (Bessel's correction).
    2. Menghitung Interquartile Range (IQR) menggunakan np.percentile.
    3. Menghitung MAD (Median Absolute Deviation) secara manual.
    4. Menghitung Range (Max - Min).
    
    Parameter:
    ----------
    data : array-like
        Kumpulan data numerik yang akan dianalisis variabilitasnya.
        
    Returns:
    -------
    metrics : dict
        Dictionary berisi nilai Standard Deviation, IQR, MAD, dan Range.
    """
    # Standard Deviation (Sample)
    std_dev = np.std(data, ddof=1)
    
    # IQR
    q75, q25 = np.percentile(data, [75, 25])
    iqr_val = q75 - q25
    
    # MAD
    median_val = np.median(data)
    mad_val = np.median(np.abs(data - median_val))
    
    # Range
    range_val = np.max(data) - np.min(data)
    
    return {
        "Standard Deviation": std_dev,
        "IQR": iqr_val,
        "MAD": mad_val,
        "Range": range_val
    }

metrics = calculate_variability_metrics(mixed_data)
print("--- Metrik Variabilitas ---")
for k, v in metrics.items():
    print(f"{k:<20}: {v:>10.4f}")

# Simulasi visualisasi Boxplot untuk variabilitas
plt.figure(figsize=(14, 6))
sns.boxplot(x=mixed_data, color="springgreen", fliersize=8, linewidth=2)
plt.title("Anatomi Boxplot: IQR, Median, dan Outliers", fontsize=16)
plt.xlabel("Nilai Data", fontsize=14)
plt.xlim(0, 600)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```

### Interpretasi Visualisasi Variabilitas

Boxplot di atas memberikan gambaran yang sangat jelas tentang anatomi dataset kita. Garis di tengah kotak mewakili **Median**. Kotak itu sendiri mencakup **Interquartile Range (IQR)**, yang berisi 50% data pusat. Perhatikan bahwa kotak ini sangat sempit dan terkonsentrasi di sekitar angka 100, menunjukkan bahwa sebagian besar data memiliki variabilitas yang rendah dan terkendali.

Garis "whiskers" yang memanjang dari kotak menunjukkan rentang data yang dianggap "normal" (biasanya 1.5 kali IQR). Titik-titik yang berada di luar garis ini secara otomatis diklasifikasikan sebagai **Outliers**. Dalam grafik kita, kita melihat deretan titik yang menjauh ke arah kanan (hingga ribuan, meskipun kita membatasi x-axis untuk kejelasan).

Pelajaran penting di sini adalah nilai **Standard Deviation** kita mungkin melonjak sangat tinggi karena titik-titik jauh tersebut, memberikan kesan seolah-olah data kita sangat tersebar (noisy). Namun, **MAD** dan **IQR** memberikan angka yang jauh lebih rendah dan realistis, mencerminkan perilaku mayoritas data. Jika Anda sedang membersihkan data untuk model regresi, memperhatikan IQR seringkali lebih bermanfaat daripada sekadar melihat standar deviasi total. Jika Standard Deviation jauh lebih besar daripada MAD, itu adalah sinyal merah (red flag) bahwa data Anda mengandung outliers ekstrem.

## 4. Eksplorasi Distribusi: Bentuk dan Karakteristik Data

Melihat angka ringkasan saja tidak cukup. Kita perlu melihat "bentuk" dari data tersebut untuk memahami proses yang menghasilkannya.

### A. Histogram dan Density Plot (KDE)
Histogram membagi data ke dalam bins. KDE (Kernel Density Estimation) memberikan kurva kontinu yang menghaluskan histogram tersebut. KDE sangat berguna untuk membandingkan beberapa distribusi dalam satu grafik.

### B. Momen Ketiga: Skewness (Kemiringan)
Mengukur asimetri distribusi. 
- **Skewness Positif**: Ekor distribusi lebih panjang di sebelah kanan (mean > median).
- **Skewness Negatif**: Ekor distribusi lebih panjang di sebelah kiri (mean < median).
$$g_1 = \frac{\frac{1}{n} \sum (x_i - \bar{x})^3}{(\frac{1}{n} \sum (x_i - \bar{x})^2)^{3/2}}$$

### C. Momen Keempat: Kurtosis (Keruncingan)
Mengukur seberapa berat ekor distribusi dibandingkan dengan distribusi normal.
- **Leptokurtik (Kurtosis Tinggi)**: Puncak tajam, ekor gemuk (banyak outlier).
- **Platikurtik (Kurtosis Rendah)**: Puncak datar, ekor tipis (sedikit outlier).

```python
def analyze_distributions(n_samples=5000):
    """
    Menghasilkan berbagai jenis distribusi statistik untuk dianalisis bentuk dan momennya.
    
    Fungsi ini mensimulasikan data dari dunia nyata:
    - Normal: Tinggi badan manusia.
    - Skewed: Pendapatan atau kekayaan.
    - Uniform: Probabilitas lemparan dadu yang adil.
    - Bimodal: Campuran dua populasi (misal: suhu siang dan malam).
    
    Parameter:
    ----------
    n_samples : int
        Jumlah data poin yang akan dihasilkan untuk setiap tipe distribusi.
        
    Returns:
    -------
    df : pandas.DataFrame
        DataFrame yang berisi kolom-kolom untuk setiap tipe distribusi.
    """
    np.random.seed(7)
    dist_data = {
        "Normal (Gausian)": np.random.normal(0, 1, n_samples),
        "Skewed (Log-Normal)": np.random.lognormal(0, 0.5, n_samples),
        "Uniform (Flat)": np.random.uniform(-2, 2, n_samples),
        "Bimodal (Mixed)": np.concatenate([np.random.normal(-2, 0.7, n_samples//2), 
                                          np.random.normal(2, 0.7, n_samples//2)])
    }
    return pd.DataFrame(dist_data)

df_dist = analyze_distributions()

fig, axes = plt.subplots(2, 2, figsize=(18, 12))
axes = axes.flatten()
colors = ["#3498db", "#e74c3c", "#2ecc71", "#f1c40f"]

for i, col in enumerate(df_dist.columns):
    sns.histplot(df_dist[col], kde=True, ax=axes[i], color=colors[i], stat="density")
    skew = stats.skew(df_dist[col])
    kurt = stats.kurtosis(df_dist[col])
    axes[i].set_title(f"{col}\nSkewness: {skew:.2f}, Kurtosis: {kurt:.2f}", fontsize=14, fontweight='bold')
    axes[i].axvline(df_dist[col].mean(), color='black', linestyle='--', label='Mean')
    axes[i].axvline(df_dist[col].median(), color='blue', linestyle='-', label='Median')
    axes[i].legend()

plt.tight_layout()
plt.suptitle("Analisis Perbandingan Berbagai Distribusi Statistik", fontsize=20, y=1.02)
plt.show()
```

### Interpretasi Visualisasi Distribusi

Empat grafik di atas menunjukkan variasi bentuk data yang sering kita temui di lapangan:

1. **Distribusi Normal**: Memiliki skewness mendekati 0 dan kurtosis mendekati 0. Ini adalah bentuk ideal "bell curve". Mean, median, dan mode berada di titik yang sama. Jika data Anda seperti ini, sebagian besar uji statistik standar akan bekerja dengan sempurna.

2. **Skewed (Log-Normal)**: Menunjukkan skewness positif. Perhatikan bagaimana Mean (garis putus-putus hitam) ditarik ke kanan menjauhi Median (garis biru). Ini sering terjadi pada data yang memiliki batas bawah nol tetapi tidak memiliki batas atas, seperti harga saham atau ukuran file.

3. **Uniform**: Memiliki kurtosis negatif yang signifikan (platikurtik). Data tersebar merata tanpa adanya pusat yang dominan. Ini menunjukkan ketidakpastian maksimum dalam rentang tertentu.

4. **Bimodal**: Menunjukkan adanya dua puncak. Ini adalah tanda peringatan bahwa data Anda mungkin berasal dari dua proses yang berbeda. Secara numerik, rata-rata mungkin berada tepat di tengah (angka nol), tetapi sebenarnya tidak ada data yang benar-benar bernilai nol. Ini disebut sebagai "rata-rata yang menyesatkan" (misleading mean).

## 5. Hubungan Antar Variabel: Korelasi dan Dependensi

Memahami bagaimana satu variabel bergerak relatif terhadap variabel lain adalah inti dari pemodelan prediktif dan feature engineering.

### A. Koefisien Pearson
Mengukur hubungan **linear**. Nilainya antara -1 hingga 1. Ia sangat peka terhadap outliers dan hanya bisa mendeteksi garis lurus.

### B. Koefisien Spearman (Rank Correlation)
Versi robust dari Pearson. Ia menggunakan peringkat data daripada nilai aslinya. Sangat baik untuk menangkap hubungan **monotonik** (selalu naik atau selalu turun) meskipun tidak linear.

### C. Koefisien Kendall's Tau
Mirip dengan Spearman tetapi lebih stabil untuk dataset kecil atau data yang memiliki banyak nilai kembar (ties).

### D. Matriks Korelasi dan Heatmap
Alat visual untuk mendeteksi multikolinearitas (ketika fitur-fitur saling berkorelasi terlalu kuat).

```python
def simulate_correlation_scenarios(n=1000):
    """
    Mensimulasikan berbagai skenario hubungan antar variabel untuk menguji koefisien korelasi.
    
    Fungsi ini dirancang untuk menunjukkan kapan Pearson gagal dan kapan Spearman unggul, 
    serta bahaya mengandalkan angka tanpa visualisasi.
    
    Langkah-langkah:
    1. Membuat variabel independen X.
    2. Membuat 4 variabel dependen Y dengan pola yang berbeda.
    3. Menghitung Pearson dan Spearman untuk setiap pasangan.
    4. Membuat scatterplot grid.
    
    Parameter:
    ----------
    n : int
        Jumlah titik data dalam setiap skenario.
    """
    x = np.linspace(0.1, 10, n)
    
    # 1. Linear Positif Kuat
    y1 = 2.5*x + np.random.normal(0, 1.5, n)
    
    # 2. Linear Negatif dengan Outliers Ekstrem
    y2 = -2*x + np.random.normal(0, 2, n)
    y2[0:20] = 100 # Suntikan outliers
    
    # 3. Non-Linear Monotonik (Logaritmik)
    y3 = np.log(x) + np.random.normal(0, 0.1, n)
    
    # 4. Hubungan Non-Monotonik (Sinusoidal)
    y4 = np.sin(x) + np.random.normal(0, 0.2, n)
    
    datasets = [y1, y2, y3, y4]
    names = ["Linear Positif", "Negatif + Outliers", "Logaritmik", "Sinusoidal"]
    
    fig, axes = plt.subplots(1, 4, figsize=(24, 6))
    
    for i, data in enumerate(datasets):
        p_corr, _ = stats.pearsonr(x, data)
        s_corr, _ = stats.spearmanr(x, data)
        
        sns.regplot(x=x, y=data, ax=axes[i], scatter_kws={'alpha':0.4, 'color':'teal'}, line_kws={'color':'red'})
        axes[i].set_title(f"{names[i]}\nPearson: {p_corr:.3f}\nSpearman: {s_corr:.3f}", fontsize=14)
        
    plt.tight_layout()
    plt.show()

simulate_correlation_scenarios()
```

### Interpretasi Visualisasi Korelasi

Hasil simulasi di atas memberikan wawasan mendalam tentang keterbatasan metrik korelasi:

1. **Linear Positif**: Kedua koefisien memberikan skor tinggi. Ini adalah kondisi ideal di mana variabel satu secara proporsional mengikuti variabel lainnya.

2. **Negatif + Outliers**: Di sini kita melihat perbedaan dramatis. **Pearson** turun signifikan karena mencoba menyesuaikan garis dengan outliers, sementara **Spearman** tetap sangat kuat karena urutan peringkat data tidak banyak berubah oleh segelintir nilai ekstrem. Ini menunjukkan Spearman lebih dapat diandalkan untuk data dunia nyata yang kotor.

3. **Logaritmik**: Hubungan ini bersifat monotonik (selalu naik) tetapi tidak linear. **Spearman** biasanya memberikan nilai yang lebih tinggi daripada Pearson di sini karena ia menangkap esensi "kenaikan bersama" tanpa terikat pada bentuk garis lurus.

4. **Sinusoidal**: Ini adalah kegagalan total bagi kedua metrik (mendekati 0). Secara visual, jelas ada hubungan yang sangat kuat, tetapi karena polanya naik dan turun, koefisien linear/monotonik tidak bisa menangkapnya. Ini menegaskan aturan emas EDA: **Selalu plot data Anda sebelum menghitung statistik!**

## 6. Teknik Deteksi Outlier Tingkat Lanjut (Multivariat)

Seringkali, sebuah titik data tidak terlihat aneh jika dilihat per kolom, tetapi menjadi sangat aneh jika dilihat dalam kombinasi beberapa kolom. Inilah yang disebut anomali multivariat.

### A. Mahalanobis Distance
Mengukur jarak titik dari pusat distribusi dalam ruang multidimensi, dengan mempertimbangkan korelasi antar variabel. Ini jauh lebih baik daripada Z-score sederhana untuk data yang berkorelasi.

### B. Isolation Forest
Algoritma berbasis tree yang "mengisolasi" observasi. Logikanya: outlier lebih mudah diisolasi (membutuhkan lebih sedikit split pada tree) daripada data normal yang padat.

```python
from sklearn.ensemble import IsolationForest

def detect_anomalies_multivariate(n_points=2000):
    """
    Mendeteksi anomali dalam ruang fitur multidimensi menggunakan Isolation Forest.
    
    Algoritma ini bekerja dengan cara membangun tree acak untuk setiap titik data. 
    Karena anomali secara intrinsik 'berbeda' dan 'langka', mereka akan berakhir 
    di cabang yang sangat pendek (mudah dipisahkan dari sisa dataset).
    
    Parameter:
    ----------
    n_points : int
        Jumlah total titik data yang dihasilkan.
    """
    # Generate data normal dengan korelasi
    rng = np.random.RandomState(42)
    X = 0.3 * rng.randn(n_points, 2)
    X_train = np.r_[X + 2, X - 2]
    
    # Tambahkan pencilan yang tersebar secara acak
    X_outliers = rng.uniform(low=-6, high=6, size=(100, 2))
    X_all = np.r_[X_train, X_outliers]
    
    # Inisialisasi dan fit model
    # Contamination didefinisikan sebagai proporsi anomali yang diharapkan
    clf = IsolationForest(contamination=0.05, random_state=42)
    y_pred = clf.fit_predict(X_all)
    
    # Plotting
    plt.figure(figsize=(12, 9))
    plt.title("Deteksi Anomali Multivariat: Isolation Forest", fontsize=16)
    
    # Pisahkan data untuk plotting
    inliers = X_all[y_pred == 1]
    outliers = X_all[y_pred == -1]
    
    plt.scatter(inliers[:, 0], inliers[:, 1], c='white', s=20, edgecolor='k', label="Inliers")
    plt.scatter(outliers[:, 0], outliers[:, 1], c='red', s=40, edgecolor='k', label="Outliers Detected")
    
    plt.legend()
    plt.xlabel("Fitur X1")
    plt.ylabel("Fitur X2")
    plt.grid(alpha=0.3)
    plt.show()

detect_anomalies_multivariate()
```

## 7. Analisis Data Kategorikal dan Tabel Kontingensi

Data kategorikal memerlukan pendekatan berbeda karena kita tidak bisa menghitung mean atau varians secara langsung.

### A. Tabel Kontingensi (Crosstabs)
Melihat frekuensi gabungan antara dua variabel kategori.

### B. Uji Chi-Square
Menilai apakah perbedaan frekuensi yang diamati signifikan secara statistik.

```python
def advanced_categorical_analysis():
    """
    Melakukan analisis mendalam pada variabel kategorikal menggunakan dataset 'tips'.
    """
    tips = sns.load_dataset("tips")
    
    # Crosstab frekuensi
    ct = pd.crosstab(tips['day'], tips['sex'], margins=True)
    print("Tabel Kontingensi Lengkap:")
    print(ct)

    # Crosstab persentase (Normalisasi)
    ct_norm = pd.crosstab(tips['day'], tips['sex'], normalize='index')
    
    # Plotting
    fig, ax = plt.subplots(1, 2, figsize=(18, 7))
    
    ct_norm.plot(kind='bar', stacked=True, ax=ax[0], color=['#ff9999','#66b3ff'])
    ax[0].set_title("Proporsi Gender per Hari", fontsize=14)
    ax[0].set_ylabel("Persentase")
    
    sns.countplot(x='day', hue='time', data=tips, ax=ax[1], palette='viridis')
    ax[1].set_title("Frekuensi Kunjungan: Siang vs Malam", fontsize=14)
    
    plt.show()

advanced_categorical_analysis()
```

## 8. Teknik Visualisasi Lanjutan untuk Data Besar

Ketika Anda memiliki jutaan baris, scatterplot biasa akan menjadi blok warna solid yang tidak informatif. Inilah cara mengatasinya:

```python
def visualize_big_data(n=50000):
    """
    Mendemonstrasikan teknik visualisasi untuk dataset besar (n=50,000).
    """
    x = np.random.normal(0, 1, n)
    y = x + np.random.normal(0, 1, n)
    
    plt.figure(figsize=(16, 6))
    
    plt.subplot(1, 2, 1)
    plt.hexbin(x, y, gridsize=30, cmap='YlOrRd')
    plt.colorbar(label='Jumlah Titik')
    plt.title("Hexbin Plot: Menangani Overplotting", fontsize=14)
    
    plt.subplot(1, 2, 2)
    sns.kdeplot(x=x, y=y, cmap='Blues', fill=True, thresh=0.05)
    plt.title("2D KDE Plot: Estimasi Kepadatan Kontinu", fontsize=14)
    
    plt.show()

visualize_big_data()
```

## 9. Statistik Matematika di Balik EDA (Deep Dive)

Sebagai koordinator Lab RnEST, Anda mungkin tertarik dengan logika formal di balik fungsi-fungsi ini.

### Derivasi Variance (Unbiased Estimator)
Kenapa kita menggunakan $n-1$ dan bukan $n$? Ini adalah **Koreksi Bessel**. Tanpa koreksi ini, rata-rata varians sampel akan lebih kecil (bias) daripada varians populasi sebenarnya.
$$E[s^2] = \sigma^2 \iff s^2 = \frac{\sum (x_i - \bar{x})^2}{n-1}$$

### Kernel Density Estimation (KDE) secara Formal
KDE bekerja dengan menempatkan fungsi kernel (biasanya Gaussian) di atas setiap titik data dan menjumlahkannya.
$$\hat{f}_h(x) = \frac{1}{nh} \sum_{i=1}^n K\left(\frac{x - x_i}{h}\right)$$
Di mana $h$ adalah **bandwidth**, parameter paling krusial yang menentukan kehalusan kurva.
