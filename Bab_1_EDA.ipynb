# Bab 1: Exploratory Data Analysis (EDA)

EDA adalah proses investigasi awal pada data untuk menemukan pola, mendeteksi anomali, menguji hipotesis, dan memeriksa asumsi dengan bantuan statistik ringkasan dan representasi grafis.

## 1. Estimasi Lokasi (Estimates of Location)
Nilai khas yang menunjukkan di mana data "berada".
- **Mean (Rata-rata)**: Sensitif terhadap outliers.
- **Median**: Nilai tengah, lebih robust (tahan) terhadap outliers.
- **Trimmed Mean**: Rata-rata setelah membuang sebagian kecil nilai ekstrem di kedua ujung.

## 2. Estimasi Variabilitas (Estimates of Variability)
Mengukur seberapa tersebar data kita.
- **Standard Deviation**: Akar kuadrat dari varians.
- **IQR (Interquartile Range)**: Selisih antara persentil ke-75 dan ke-25. Sangat bagus untuk mendeteksi outliers secara statistik.

```python
import pandas as pd
import numpy as np

data = [1, 2, 3, 4, 5, 100] # 100 adalah outlier

print(f"Mean: {np.mean(data)}")
print(f"Median: {np.median(data)}")
```

## 3. Visualisasi Distribusi
- **Boxplot**: Menunjukkan median, kuartil, dan outliers secara visual.
- **Histogram**: Menunjukkan frekuensi data dalam rentang tertentu.

## 4. Korelasi
Mengukur hubungan antara dua variabel. Koefisien korelasi Pearson berkisar antara -1 hingga 1.
- 1: Korelasi positif sempurna.
- 0: Tidak ada korelasi linear.
- -1: Korelasi negatif sempurna.
